```r
source(file = "util.r")

install_packages(c('ggplot2', 'zoo', 'forecast', 'quantmod'), quietly = T, warn.conflicts = F)

library(repr)
options(repr.plot.width=6, repr.plot.height=3)
```

### Trend

```r
data.ts = get_bond_yields_uk_10y_data_ts('quarterly_csv.csv')
autoplot(data.ts) +
  ggtitle("10y UK Government Bond Yields (long-term interest rate)") +
  xlab("Year") +
  ylab("Rate")
```

### Cycle and Trend

```r
data.ts = get_stedmondsbury_new_homes_ts()
autoplot(data.ts) +
  ggtitle("St Edmundsbury new homes") +
  xlab("Year") +
  ylab("Number sold")
```

### Seasonal and Trend

```r
autoplot(AirPassengers) +
  ggtitle("AirPassengers") +
  xlab("Year") +
  ylab("# Travellers (000s)")
```
