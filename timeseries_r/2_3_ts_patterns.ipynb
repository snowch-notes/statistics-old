```r
source(file = "util.r")

install_packages(c('ggplot2', 'zoo', 'forecast', 'quantmod', 'gridExtra'), quietly = T, warn.conflicts = F)

library(repr)
options(repr.plot.width=12, repr.plot.height=6)
```

```r
data.ts = get_bond_yields_uk_10y_data_ts('quarterly_csv.csv')
plot1 = autoplot(data.ts) +
  ggtitle("[Trend] 10y UK Bond Yields") +
  xlab("Year") +
  ylab("Rate")

data.ts = get_stedmondsbury_new_homes_ts()
plot2 = autoplot(data.ts) +
  ggtitle("[Cycle and Trend] St Edmundsbury new homes") +
  xlab("Year") +
  ylab("Number sold")

plot3 = autoplot(AirPassengers) +
  ggtitle("[Seasonal and Trend] AirPassengers") +
  xlab("Year") +
  ylab("# Travellers (000s)")

daily_change = get_ibm_daily_close_change_ts()
plot4 = autoplot(daily_change) +
  ggtitle("[Random] Daily Change in IBM Closing Price") +
  xlab("Year") +
  ylab("delta")

grid.arrange(plot1, plot3, plot2, plot4, nrow=2, ncol=2)
```
