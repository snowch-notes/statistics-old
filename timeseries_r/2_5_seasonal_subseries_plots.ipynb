```r
source(file = "util.r")

install_packages(c('ggplot2', 'zoo', 'forecast', 'quantmod'), quietly = T, warn.conflicts = F)

library(repr)
options(repr.plot.width=6, repr.plot.height=3)
```

### Seasonal and Trend

```r
autoplot(AirPassengers) +
  ggtitle("AirPassengers") +
  xlab("Year") +
  ylab("# Travellers (000s)")
```

```r
ggseasonplot(AirPassengers, year.labels=TRUE, year.labels.left=TRUE) +
  ylab("# Travellers (000s)") +
  ggtitle("AirPassengers")
```

```r
ggsubseriesplot(AirPassengers, year.labels=TRUE, year.labels.left=TRUE, polar = TRUE) +
  ylab("# Travellers (000s)") +
  ggtitle("AirPassengers")
```
